```python
from difflib import SequenceMatcher
from IPython.display import HTML, display

def levenshtein(s1, s2):
    if len(s1) < len(s2):
        return levenshtein(s2, s1)
    if len(s2) == 0:
        return len(s1)
    prev_row = list(range(len(s2) + 1))
    for i, c1 in enumerate(s1):
        curr_row = [i + 1]
        for j, c2 in enumerate(s2):
            ins = prev_row[j + 1] + 1
            dele = curr_row[j] + 1
            sub = prev_row[j] + (c1 != c2)
            curr_row.append(min(ins, dele, sub))
        prev_row = curr_row
    return prev_row[-1]

def cer(ref, hyp):
    ref = ref.strip()
    hyp = hyp.strip()
    d = levenshtein(ref, hyp)
    return d / max(len(ref), 1)

def wer(ref, hyp):
    ref_words = ref.strip().split()
    hyp_words = hyp.strip().split()
    d = levenshtein(ref_words, hyp_words)
    return d / max(len(ref_words), 1)

def show_word_diff(ref, hyp):
    ref = ' '.join(ref.strip().split())
    hyp = ' '.join(hyp.strip().split())
    ref_words = ref.split()
    hyp_words = hyp.split()
    matcher = SequenceMatcher(None, ref_words, hyp_words)
    html = '<div style="font-family: monospace; white-space: pre-wrap;">'
    for tag, i1, i2, j1, j2 in matcher.get_opcodes():
        if tag == 'equal':
            html += ' '.join(ref_words[i1:i2]) + ' '
        elif tag == 'replace':
            html += '<span style="color:red; text-decoration: line-through;">' + ' '.join(ref_words[i1:i2]) + '</span> '
            html += '<span style="color:red;">' + ' '.join(hyp_words[j1:j2]) + '</span> '
        elif tag == 'delete':
            html += '<span style="color:red; text-decoration: line-through;">' + ' '.join(ref_words[i1:i2]) + '</span> '
        elif tag == 'insert':
            html += '<span style="color:red;">' + ' '.join(hyp_words[j1:j2]) + '</span> '
    html += '</div>'
    display(HTML(html))
```
